```python
import json
from pathlib import Path
import pandas as pd

DATA_PATH = Path("..") / "Resume Data" / "resumes-original.json"
assert DATA_PATH.exists(), f"Missing file: {DATA_PATH}"
```

```python
data = json.loads(DATA_PATH.read_text())
resumes = data["resumes"]
len(resumes), resumes[0].keys()
```

```python
def resume_to_text(r: dict) -> str:
    college = r.get("college", {})
    academics = r.get("academics", {})
    skills = r.get("skills", {})
    exp = r.get("experience", {})

    major = college.get("major", "")
    school = college.get("name", "")
    degree = college.get("degree", "")
    cgpa = academics.get("cgpa", None)

    courses = academics.get("relevant_coursework", [])
    course_text = " | ".join(
        f"{c.get('course','')} ({c.get('term','')}): {c.get('description','')}".strip()
        for c in courses
    )

    tech = skills.get("technical", [])
    soft = skills.get("soft", [])

    projects = exp.get("projects", [])
    proj_text = " | ".join(
        f"{p.get('name','')} — {p.get('role','')}. {p.get('description','')}".strip()
        for p in projects
    )

    work = exp.get("work", [])
    work_text = " | ".join(
        f"{w.get('title','')} at {w.get('company','')}. {w.get('description','')}".strip()
        for w in work
    )

    leadership = exp.get("leadership_and_activities", [])
    lead_text = " | ".join(
        f"{a.get('role','')} ({a.get('activity','')}): {a.get('description','')}".strip()
        for a in leadership
    )

    return "\n".join([
        f"Major: {major}",
        f"School: {school} | Degree: {degree}",
        f"CGPA: {cgpa}",
        f"Technical skills: {', '.join(tech)}",
        f"Soft skills: {', '.join(soft)}",
        f"Coursework: {course_text}",
        f"Projects: {proj_text}",
        f"Work: {work_text}",
        f"Leadership: {lead_text}",
    ]).strip()

texts = [resume_to_text(r) for r in resumes]
texts[0][:400]
```

```python
import sys
print(sys.executable)
```

```python

from sentence_transformers import SentenceTransformer

model = SentenceTransformer("paraphrase-MiniLM-L6-v2")
embeddings = model.encode(texts, show_progress_bar=True, normalize_embeddings=True)
embeddings.shape
```

```python
import chromadb
from chromadb.config import Settings

CHROMA_DIR = Path("chroma_store")
client = chromadb.PersistentClient(path=str(CHROMA_DIR))

# reset collection for reruns
try:
    client.delete_collection("resumes")
except:
    pass

collection = client.create_collection(name="resumes")
```

```python
def metadata_from_resume(r: dict) -> dict:
    college = r.get("college", {})
    academics = r.get("academics", {})
    skills = r.get("skills", {})
    exp = r.get("experience", {})

    return {
        "student_key": r.get("student_key"),
        "major": college.get("major"),
        "school": college.get("name"),
        "cgpa": float(academics.get("cgpa")) if academics.get("cgpa") is not None else None,
        "n_tech": len(skills.get("technical", [])),
        "n_soft": len(skills.get("soft", [])),
        "n_projects": len(exp.get("projects", [])),
        "n_work": len(exp.get("work", [])),
    }

ids = [r["student_key"] for r in resumes]
metas = [metadata_from_resume(r) for r in resumes]

collection.add(
    ids=ids,
    documents=texts,
    embeddings=embeddings.tolist(),
    metadatas=metas,
)

collection.count()
```

```python
def query_similar(student_key: str, k: int = 5, filter_major: bool = False):
    r = next(x for x in resumes if x["student_key"] == student_key)
    qtext = resume_to_text(r)
    qemb = model.encode([qtext], normalize_embeddings=True).tolist()

    where = None
    if filter_major:
        where = {"major": r["college"]["major"]}

    res = collection.query(
        query_embeddings=qemb,
        n_results=k,
        where=where
    )
    return res

result = query_similar("u_0007", k=5, filter_major=False)
result.keys()
```

```python
def format_results(res):
    out = []
    for i in range(len(res["ids"][0])):
        out.append({
            "rank": i+1,
            "student_key": res["ids"][0][i],
            "distance": res["distances"][0][i],
            "major": res["metadatas"][0][i].get("major"),
            "cgpa": res["metadatas"][0][i].get("cgpa"),
            "n_projects": res["metadatas"][0][i].get("n_projects"),
            "n_tech": res["metadatas"][0][i].get("n_tech"),
        })
    return pd.DataFrame(out)

format_results(result)
```

```python
def get_skills_set(r):
    tech = set(r.get("skills", {}).get("technical", []))
    soft = set(r.get("skills", {}).get("soft", []))
    return tech, soft

def suggest_missing_skills(query_key: str, k: int = 10):
    q = next(x for x in resumes if x["student_key"] == query_key)
    qtech, qsoft = get_skills_set(q)

    res = query_similar(query_key, k=k, filter_major=False)
    neighbor_ids = [sid for sid in res["ids"][0] if sid != query_key]

    tech_counts = {}
    soft_counts = {}

    for nid in neighbor_ids:
        nr = next(x for x in resumes if x["student_key"] == nid)
        ntech, nsoft = get_skills_set(nr)

        for s in ntech:
            tech_counts[s] = tech_counts.get(s, 0) + 1
        for s in nsoft:
            soft_counts[s] = soft_counts.get(s, 0) + 1

    # skills common in neighbors but not in me
    tech_missing = sorted(
        [(s,c) for s,c in tech_counts.items() if s not in qtech],
        key=lambda x: (-x[1], x[0])
    )[:5]

    soft_missing = sorted(
        [(s,c) for s,c in soft_counts.items() if s not in qsoft],
        key=lambda x: (-x[1], x[0])
    )[:5]

    return {"tech_missing_top": tech_missing, "soft_missing_top": soft_missing}

suggest_missing_skills("u_0007", k=10)
```
